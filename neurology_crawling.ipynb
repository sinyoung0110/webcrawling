```python
# 셀레니움 기본 템플릿
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


# 크롬 드라이버 생성
driver = webdriver.Chrome()
```

```python
# 페이지 이동
driver.get("https://www.neurology.org/all-articles")
```

```python
# 태그 찾기 (제일 큰 단위 박스)
articles = driver.find_elements(By.CSS_SELECTOR, ".card.border-bottom")

# 태그 찾기 (제목선택자)
# articles = driver.find_elements(By.CSS_SELECTOR, '.card.border-bottom h3 > a')

# 제목을 가져올 수는 있는데, 제목만 가져온다 -> .card.border-bottom 에서 제목, abstract, URL, 가져와야 하는데.. 어떻게 해야 하지??
```

```python
# 페이지 로드 대기
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '.card.border-bottom')))

import time
import pandas as pd

articles_data = []

for article in articles:
    title_element = article.find_element(By.CSS_SELECTOR, "h3 > a")
    title = title_element.text.strip()
    date_element = article.find_element(By.CSS_SELECTOR, ".card.border-bottom .overline")
    date = date_element.text.strip()
    link_element = article.find_element(By.CSS_SELECTOR, ".card-footer .card-formats > a")
    link_href = link_element.get_attribute("href")
    abstract_element = article.find_element(By.CSS_SELECTOR, ".card-body")
    abstract = abstract_element.text
    print(title, date, link_href, abstract)
    articles_data.append((title, date, link_href, abstract))

# link (abstrac link)
# author
# abstract
```

```python
# 데이터 프레임 생성
df = pd.DataFrame(articles_data, columns=['제목', '날짜', '링크', '초록'])
df
```
